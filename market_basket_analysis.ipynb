```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
orders = pd.read_csv("Data/orders.csv")
order_products_prior =pd.read_csv("Data/order_products__prior.csv")
order_products_train = pd.read_csv("Data/order_products__train.csv")
products = pd.read_csv("Data/products.csv")
```

```python
order_products = order_products_prior.append(order_products_train)
order_products.shape
```

```python
order_products.head()
```

```python
order_products.product_id.nunique()
```

```python
product_counts = order_products.groupby('product_id')['order_id'].count().reset_index().rename(columns = {'order_id':'frequency'})
product_counts = product_counts.sort_values('frequency', ascending=False)[0:49688].reset_index(drop = True)
product_counts = product_counts.merge(products, on = 'product_id', how = 'left')
product_counts.head(10)
```

```python
product_counts
```

```python
freq_products = list(product_counts.product_id)
freq_products[1:10]
```

```python
freq_products
len(freq_products)
```

```python
order_products = order_products[order_products.product_id.isin(freq_products)]
order_products.shape
```

```python
order_products.order_id.nunique()
```

```python
order_products = order_products.merge(products, on = 'product_id', how='left')
order_products.head()
```

```python
order_products.shape
```

```python
basket = order_products.groupby(['order_id', 'product_name'])['reordered'].count().unstack().reset_index().fillna(0).set_index('order_id')
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1 
basket = basket.applymap(encode_units)
basket.head()
```

```python
basket.size
```

```python
basket.shape
```

```python
frequent_items = apriori(basket, min_support=0.01, use_colnames=True, low_memory=True)
frequent_items.head()
```

```python
frequent_items.tail()
```

```python
frequent_items.shape
```

```python
rules = association_rules(frequent_items, metric="lift", min_threshold=1)
data = rules.sort_values('lift', ascending=False)
```

```python
data.shape
```

```python
def predict(inp):
    res = data.loc[data['antecedents'] == inp]
    print(res, type(res))
predict('Banana')
```

```python
res = data[data['support'] == 0.01186]
print(res)
```
